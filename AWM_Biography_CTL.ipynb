## II. Creation Process
### A. Framing the Problem
Internal legal departments often outsource legalwork to outside firms. In order to outsource legal work, a company may require Request for Outside Counsel Approval, or ROCA, which is approved by an Attorney in a legal department. 

For a large company, these requests can come at an extremely high volume.  Many times, teams are inundated with these requests to the extent that it effects their work product on other tasks.  In order to combat this, companies make it a priority to change processes and increasing efficiencies.  This is where my problem truly arises. Because of the vast quantity of these requests, Attorneys need a system that helps them increase the speed of business and decrease the amount of time spent reviewing and researching firms.  To do so, I considered ways that I could streamline their decision making. 


### B. Existing Solutions

This solution was built out by taking agile workflows and combining them with design thinking ideology.  To do so, the team spent many hours interviewing different groups to identify the workflow for this process. Eventually, working with each practice group, we were able to quantify these criteria and develop standard work that changed this process. This existing solution was a viable way to decrease the volume of requests, even if only marginally.  However, the larger issue was still prevalent; Attorneys have a higher volume of requests than they can handle manually.

### B. Impact & Efficiencies
This tool offers a prospect of greatly increasing the efficiency of attorneys that are reviewing ROCAs. The tool can offer this because by having a reference guide that suggests whether a matter should be approved cuts down on the one of the most time consuming steps in the ROCA process: understanding the firm rates in the context in which they apply.  For example, in the traditional work flow, an attorney would have to do research from the ground up, determining what like firms charge in a particular practice area, and then comparing that to the rates on the request.  Furthermore, the attorney would need to do extensive research on the firm themselves, including going to contacts within the company to see if anyone could recommend a firm.  By using our tool however, they can cut down on the time they do this research, because they would know that the algorithm has considered these extrinsic factors. By doing so they increase the speed of business because they spend less time doing manual research.

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

from sklearn.metrics import roc_curve, auc, precision_recall_curve
import matplotlib.pyplot as plt

def evaluate(prob, pred, labels_test, verbose=1):
    acc = accuracy_score(labels_test,pred)
    ones = sum(labels_test)/len(labels_test)
    zeros = (1 - ones)
    tn, fp, fn, tp = confusion_matrix(labels_test, pred).ravel()

    recall_ = tp / (tp + fn)
    precision_ = tp / (tp + fp)
    f1 = (2 / ((1/recall_)+(1/precision_)))

    false_positive_rate, true_positive_rate, thresholds = roc_curve(labels_test,prob)
    roc_auc = auc(false_positive_rate, true_positive_rate)

    if verbose==1:

        plt.rcParams['figure.figsize'] = [14, 4]

        plt.subplot(1, 3, 1)
        plt.title('Receiver Operating Characteristic')
        plt.plot(false_positive_rate, true_positive_rate, 'b',
        label='AUC = %0.2f'% roc_auc)
        plt.legend(loc='lower right')
        plt.plot([0,1],[0,1],'r--')
        plt.xlim([-0.1,1.2])
        plt.ylim([-0.1,1.2])
        plt.ylabel('True Positive Rate')
        plt.xlabel('False Positive Rate')

        precision, recall, thresholds = precision_recall_curve(labels_test,prob)

        plt.subplot(1, 3, 2)
        plt.title('Precision-Recall Curve')
        plt.plot(recall, precision,'b',
        label='AP = %0.2f'% precision.mean())
        plt.legend(loc='lower right')
        plt.xlim([-0.1,1.2])
        plt.ylim([-0.1,1.2])
        plt.ylabel('Recall')
        plt.xlabel('Precision')

        plt.subplot(1, 3, 3)
        plt.hist([labels_test,prob], label=["true","pred"], bins='auto')  # arguments are passed to np.histogram
        plt.title("Predictions vs Reality")
        plt.legend(loc='best')
        plt.xlim([-0.1,1.2])

        plt.tight_layout()
        plt.show()

        print ("\n0s: %0.2f\tTrue Positives: %s\tAccuracy: %s "%(zeros,tp,acc))
        print ("1s: %0.2f\tTrue Negatives: %s\tAUC: %s"%(ones,tn,roc_auc))
        print ("\t\tFalse Positives: %s\tF1 Score: %s"%(fp,f1))
        print ("\t\tFalse Negatives: %s\tRecall (fract of actual yeas found): %s"%(fn,recall_))
        print ("\t\t\t\t\tPrecision (correctness of yeas predicted): %s\n"%precision_)

        null = zeros
        if null < ones:
            null = ones

        if (acc>null) and (roc_auc>0.5) and (recall_>0.5) and (precision_>0.5):
            print("################")
            print("#     PASS     #")
            print("################")
            vpass = "Yes"
        else:
            print("################")
            print("#     FAIL     #")
            print("################")
            vpass = "No"
```

```python
class_df = final_table.copy()
class_df.head()
```

```python
# create a dataframe containing a random sample of rows
class_holdout = class_df.sample(frac=0.20)

# create a dataframe that conatins the rows from except for those in holdout
class_training = class_df.loc[~class_df.index.isin(class_holdout.index)]

# make a training dataframe containing just features
features_train = class_training.drop("Accept", axis=1).as_matrix(columns=None)

# make a training dataframe containing only the target
labels_train = class_training["Accept"].as_matrix(columns=None)

# make a testing/holdout dataframe containing just features
features_test = class_holdout.drop("Accept", axis=1).as_matrix(columns=None)

# make a testing/holdout dataframe containing only the target
labels_test = class_holdout["Accept"].as_matrix(columns=None)

print("size of training",len(class_training))
```

```python
from sklearn.linear_model import LogisticRegression
# Logistic Regression
model = LogisticRegression(fit_intercept = False, C = 1e9)
clf_1 = model.fit(features_train, labels_train)
pred = clf_1.predict(features_test)
prob = clf_1.predict_proba(features_test)[:,1] 
print("Logistic Regression")
evaluate(prob, pred, labels_test)
```

```python
from sklearn import tree
clf_2 = tree.DecisionTreeClassifier(min_samples_split=40)
clf_2 = clf_2.fit(features_train, labels_train)
pred = clf_2.predict(features_test)
prob = clf_2.predict_proba(features_test)[:,1] 
print("\nDecision Tree")
evaluate(prob, pred, labels_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf_3 = RandomForestClassifier()
clf_3 = clf_3.fit(features_train, labels_train)
pred = clf_3.predict(features_test)
prob = clf_3.predict_proba(features_test)[:,1] 
print("Random Forest")
evaluate(prob,pred, labels_test)
```
